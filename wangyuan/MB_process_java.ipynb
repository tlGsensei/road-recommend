```python
import pandas as pd
data = pd.read_csv('./processed/java_nodul.csv',sep='\t')
data
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
data[['uid','sid']] = data[['stuno','title']].apply(encoder.fit_transform)
data
```

```python
data['uid'] = data['uid'].apply(lambda x: x+1)
data['sid'] = data['sid'].apply(lambda x: x+1)
data
```

```python
res = data[data['new_oper']!='duration'][['uid', 'sid', 'new_oper', 'time']]
res
```

```python
res.to_csv('./processed/mb_java.csv', sep='\t', index=False, header=None)
```
