```python
processed = data[~data['title'].isnull()]
processed.sort_values(by=['stuno', 'time'], ascending=True, inplace=True)
processed
```

```python
def change(x):
    if x in [60,65,67,72,1]:
        return 'watch'
    elif x in [4,78,79,77,58]:
        return 'tag'
    elif x in [75,71,56,57]:
        return 'done'
    elif x in [62,64,68,69,70]:
        return 'enter'
    else:
        return 'duration'
processed['new_oper'] = processed['oper'].apply(lambda x: change(x))
processed
```
