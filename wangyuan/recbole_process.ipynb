```python
import pandas as pd
java = pd.read_csv('./processed/java_mb.txt',sep='\t',names=['user_id:token','item_id:token','behavior','timestamp:float'])
os = pd.read_csv('./processed/os_mb.txt',sep='\t',names=['user_id:token','item_id:token','behavior','timestamp:float'])
net = pd.read_csv('./processed/net_mb.txt',sep='\t',names=['user_id:token','item_id:token','behavior','timestamp:float'])
java
```

```python
java[['user_id:token','item_id:token','timestamp:float']].to_csv('./processed/java.inter',sep='\t',header=True,index=False)
os[['user_id:token','item_id:token','timestamp:float']].to_csv('./processed/os.inter',sep='\t',header=True,index=False)
net[['user_id:token','item_id:token','timestamp:float']].to_csv('./processed/net.inter',sep='\t',header=True,index=False)
```

```python
def change(x):
    if x == 'watch':
        return 0
    elif x == 'done':
        return 1
    elif x == 'tag':
        return 2
    elif x == 'enter':
        return 3
    else:
        return 4
java_mb = java
java_mb['item_type'] = java_mb['behavior'].apply(lambda x : change(x))
java_mb
```

```python
java_inter = java_mb.groupby('user_id:token').agg({'item_id:token':'unique'}).reset_index()
java_beh = java_mb.groupby('user_id:token').agg({'item_type':'unique'}).reset_index()
java_inter
```

```python
java_inter['item_id_list:token_seq'] = java_inter['item_id:token'].apply(lambda x: ' '.join([str(i) for i in x]))
java_inter
```

```python
java_beh['item_type_list:token_seq']=java_beh['item_type'].apply(lambda x: ' '.join([str(i) for i in x]))
java_beh
```

```python
java_inter['item_id:token']=java_inter['item_id:token'].apply(lambda x: x[-1])
java_inter
```

```python
java2 = pd.merge(java_beh[['user_id:token','item_type_list:token_seq']],java_inter[['user_id:token','item_id_list:token_seq','item_id:token']],how='inner',on='user_id:token')
java2
```

```python
# sample for test and train
java_train = java2.sample(frac=0.8)
java_test = java2[~java2.index.isin(java_train.index)]
print(len(java_train),len(java_test))
```

```python
java_test['session_id:token']=java_test['user_id:token']
java_train['session_id:token']=java_train['user_id:token']
java_test[['session_id:token','item_id_list:token_seq','item_type_list:token_seq','item_id:token']].to_csv('java.test.inter',sep='\t',header=True,index=False)
java_train[['session_id:token','item_id_list:token_seq','item_type_list:token_seq','item_id:token']].to_csv('java.train.inter',sep='\t',header=True,index=False)
```
