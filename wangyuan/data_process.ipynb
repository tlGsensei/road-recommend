```python
import pandas as pd
import numpy as np
data = pd.read_csv('wangyuan_original.csv',sep=',')
data
```

```python
data['oper'].value_counts()
```

```python
def change_behave(x):
    if x in [4,75,67,69,17,60,1,65,72,57]:
        return "learn"
    elif x in [78,70]:
        return "pause"
    elif x in [77]:
        return "continue"
    elif x in [63,64,66,68,55,74,62]:
        return "enter"
    elif x in [71,58,56]:
        return "submit"
    elif x in [79]:
        return "drag"
```

```python
data['behavior'] = data['oper'].apply(lambda x: change_behave(x))
data
```

```python
processed_data = data.dropna()
processed_data
```

```python
processed_data['behavior'].value_counts()
```

```python
wy = processed_data[['stuno','title','behavior','rtime']]
wy
```

```python
import time
wy['timestamp'] = wy['rtime'].apply(lambda x:time.mktime(time.strptime(x,'%d/%m/%Y %H:%M:%S')))
wy
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
wy[['uid','sid']] = wy[['stuno','title']].apply(encoder.fit_transform)
wy
```

```python
final_wy = wy[['uid','sid','behavior','timestamp']]
final_wy
```

```python
final_wy['sid'].nunique()
```

```python
final_wy['uid'].nunique()
```

```python
final_wy['behavior'].value_counts()
```

```python
final_wy.to_csv('wangyuan.csv',sep='\t',index=False,header=None)
```

```python
def trans(x):
    if x in ['drag','continue','pause']:
        return "tip"
    elif x =="enter":
        return "netural"
    elif x == "learn":
        return "pos"
    else:
        return "neg"
```

```python
final_wy['behavior']=final_wy['behavior'].apply(lambda x:trans(x))
final_wy
```

```python
final_wy['behavior'].value_counts()
```

```python
final_wy.to_csv("wangyuan.txt",sep='\t',header=None,index=False)
```
