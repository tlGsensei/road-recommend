```python
#读入数据
import pandas as pd
data = pd.read_csv('wangyuan_original.csv')

import time
data['time'] = data['rtime'].apply(lambda x: int(time.mktime(time.strptime(x, "%d/%m/%Y %H:%M:%S"))))

processed = data[~data['title'].isnull()]
processed.sort_values(by=['stuno', 'time'], ascending=True, inplace=True)

def change(x):
    if x in [60,65,67,72,1]:
        return 'watch'
    elif x in [4,78,79,77,58]:
        return 'tag'
    elif x in [75,71,56,57]:
        return 'done'
    elif x in [62,64,68,69,70]:
        return 'enter'
    else:
        return 'duration'
processed['new_oper'] = processed['oper'].apply(lambda x: change(x))
processed
```

```python
# 获取处理后的java数据
java = processed[processed['courseid']==5]
java = java[['stuno','courseid','oper','title','new_oper','time','tlen']]
no_hm = java[~java['title'].str.contains('作业')]
no_hm.drop_duplicates(subset=['stuno','title','new_oper'], keep='last', inplace=True)
no_hm_count = no_hm.groupby('stuno').count().sort_values(by = ['oper'], ascending=False)
pro_java = no_hm[no_hm['stuno'].isin(no_hm_count[no_hm_count['oper']>=4].index)]
pro_java
```

删除重复的动作记录，java剩下2178条记录可用, 操作系统剩下4045条可用,计算机网络剩下3087条可用

```python
# 获取处理后的os数据
os = processed[processed['courseid']==23]
os = os[['stuno','courseid','oper','title','new_oper','time','tlen']]
no_hm_os = os[~os['title'].str.contains('作业')]
no_hm_os.drop_duplicates(subset=['stuno','title','new_oper'], keep='last', inplace=True)
no_hm_os_count = no_hm_os.groupby('stuno').count().sort_values(by = ['oper'], ascending=False)
pro_os = no_hm_os[no_hm_os['stuno'].isin(no_hm_os_count[no_hm_os_count['oper']>=4].index)]
pro_os
```

```python
# 获取处理后的net数据
net = processed[processed['courseid']==182]
net = net[['stuno','courseid','oper','title','new_oper','time','tlen']]
no_hm_net = net[~net['title'].str.contains('作业')]
no_hm_net.drop_duplicates(subset=['stuno','title','new_oper'], keep='last', inplace=True)
no_hm_net_count = no_hm_net.groupby('stuno').count().sort_values(by = ['oper'], ascending=False)
pro_net = no_hm_net[no_hm_net['stuno'].isin(no_hm_net_count[no_hm_net_count['oper']>=4].index)]
pro_net
```
