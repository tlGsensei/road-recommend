```python
import pandas as pd
java = pd.read_csv('./processed/java_nodul.csv',header=0,sep='\t')
java
```

```python
java_count = java.groupby('stuno').count()
pro_java = java[java['stuno'].isin(java_count[java_count['oper']>3].index)][['stuno','title','new_oper','time']]
pro_java
```

```python
java_dict = pd.read_csv('./java/entity2id.txt',sep='\t', names=['entity','id'])
java_dict
```

```python
java_dict = java_dict.set_index('entity')['id'].to_dict()
```

```python
pro_java['new_title'] = pro_java['title'].apply(lambda x: True if java_dict.get(x) else False)
pro_java
```

```python
pro_java[pro_java['new_title']==False]['title'].nunique()
```

```python
pro_java[pro_java['new_title']==True]['title'].nunique()
```

```python
not_in_java = pro_java[pro_java['new_title']==False]
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
not_in_java[['new_title2']] = not_in_java[['title']].apply(encoder.fit_transform)
not_in_java
```

```python
not_in_java['new_title2'] = not_in_java['new_title2']+363
not_in_java
```

```python
in_java = pro_java[pro_java['new_title']==True]
in_java['new_title2'] = in_java['title'].apply(lambda x: java_dict.get(x))
in_java
```

```python
new_java = pd.concat([in_java, not_in_java])
new_java.sort_values(by=['stuno','time'])
new_java
```

```python
new_java[['new_stuno']] = new_java[['stuno']].apply(encoder.fit_transform)
```

```python
new_java2 = new_java[['new_stuno','new_title2','new_oper','time']]
new_java2['new_stuno'] = new_java2['new_stuno'] + 1
new_java2
```

```python
new_java2.to_csv('./processed/java_mb.txt',header=None, sep='\t', index=False)
```

```python
new_java2['new_title2'].max()
```

```python
max(new_java2.groupby('new_stuno').count()['time'])
```

```python
max(new_java2['new_title2'])
```
