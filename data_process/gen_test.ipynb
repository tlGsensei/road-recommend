```python
import pandas as pd
java = pd.read_csv('java_mb_KG.txt',names=['user_id', 'item_id','behavior','timestamp'],sep='\t')
java
```

```python
import random
java_user = list(set(java['user_id'].tolist()))
sample_num = int((len(java_user)+1)*0.9)
java_user_train = random.sample(java_user, sample_num)
train_java = java[java['user_id'].isin(java_user_train)]
test_java = java[~java['user_id'].isin(java_user_train)]
```

```python
train_java
```

```python
test_java
```

```python
train_java.to_csv('java_mb_KG_train.txt',sep='\t',header=False,index=False)
test_java.to_csv('java_mb_KG_test.txt',sep='\t',index=False,header=False)
```
