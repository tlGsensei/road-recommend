```python
def cal_f(x,y):
    return 2*x*y/(x+y)
```

```python
import pandas as pd
data = []
for i in range(5):
    for _ in range(10):
        tmp = []
        tmp.append(i+1)
        tmp.append(p_1[i]+np.random.normal(scale=0.05,loc=0))
        tmp.append(r_1[i]+np.random.normal(scale=0.005,loc=0))
        tmp.append(cal_f(tmp[1],tmp[2]))
        data.append(tmp)

df = pd.DataFrame(data, columns=['path_length','precision','recall','f1_score'])
df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.boxplot(x="path_length", y="precision", data=df, palette="Set1", width=0.5)
plt.savefig("./c_class_p.png")
plt.show()
```
